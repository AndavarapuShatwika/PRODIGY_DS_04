```python
duplicates = twitter.duplicated()
duplicated_rows = twitter[duplicates]
duplicated_rows.count()
```

```python
# Dropping Duplicates
twitter.drop_duplicates(inplace=True)
```

```python
# Number of Unique Values
twitter.nunique()
```

```python
twitter['Positive'].value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.pie(x=twitter['Positive'].value_counts().values, 
        labels=twitter['Positive'].value_counts().index, 
        autopct='%.1f%%', explode=[0.03, 0.03,0.03,0.03])
plt.title('The Distribution of Sentiment')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(x='Borderlands', hue='Positive', data=twitter)
plt.title('Sentiment Distribution by Entity')
plt.xticks(rotation=45, fontsize=5)
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x=twitter['Borderlands'].value_counts().values,y=twitter['Borderlands'].value_counts().index)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Entity Distribution')
plt.show()
```
